```python
import pandas as pd
import json
import os.path
import matplotlib.pyplot as plt
import collections
import matplotlib as m
import copy
```

```python
df = pd.read_csv("Permissions_all.csv", index_col = 0)
```

```python
for i,perm in enumerate(df["Permissions"]):
    if type(perm) == str:
        df.at[i,"Permissions"] = json.loads(perm)
```

```python

buckets = collections.defaultdict(collections.Counter)

for perm in df["Permissions"]:
    if type(perm) == dict:
        for _type, value in perm.items():
            buckets[_type].update(value)
```

```python
buckets["Photo_Files"] = buckets['Photos/Media/Files']
del buckets['Photos/Media/Files']

buckets["Device_ID_&_Call_Info"] = buckets['Device ID & call information']
del buckets['Device ID & call information']

buckets["Wifi"] = buckets['Wi-Fi connection information']
del buckets['Wi-Fi connection information']

buckets["Device_&_App_History"] = buckets['Device & app history']
del buckets['Device & app history']

buckets['Wearable_Sensors_Data'] = buckets['Wearable sensors/Activity data']
del buckets['Wearable sensors/Activity data']

buckets['Cellular_Data_Settings'] = buckets['Cellular data settings']
del buckets['Cellular data settings']

#dict_keys(['Camera', 'Other', 'Identity', 'Contacts', 'Location', 'Phone', 'Photos/Media/Files', 'Storage', 'Microphone', 'Device ID & call information', 'Wi-Fi connection information', 'Uncategorized', 'Calendar', 'Device & app history', 'SMS', 'Wearable sensors/Activity data', 'Cellular data settings'])
```

```python


print("Photo_Files".replace("_", " "))
```

```python
size = len(df["Permissions"])
for key,val in buckets.items():
    if "Photo_Files" == key:
        plt.bar(val.keys(), val.values())
        plt.title(key)
        plt.xticks(rotation = 60)
        plt.savefig(fname = "photosmediafiles.png", format = "png",bbox_inches = 'tight')
        plt.show()
    elif key == "Other":
        n_val = {k:v for k, v in val.items() if v > 400000}
        plt.bar(n_val.keys(), n_val.values())
        plt.title(key)
        plt.xticks(rotation = 60)
        plt.savefig(fname = key + ".png", format = "png", bbox_inches = 'tight')
        plt.show()
    elif key == "Uncategorized":
        n_val = {k:v for k, v in val.items() if v > 10000}
        plt.bar(n_val.keys(), n_val.values())
        plt.title(key)
        plt.xticks(rotation = 60)
        plt.savefig(fname = key + ".png", format = "png", bbox_inches = 'tight')
        plt.show()
    else:
        percent = [v for v in val.values()]
            
        plt.bar(val.keys(), val.values())
        plt.title(key)
        plt.xticks(rotation = 60)
        plt.savefig(fname = key.replace("_", " ") + ".png", format = "png", bbox_inches = 'tight')
        plt.show()
```

```python

total_counts = {perm:sum(val.values()) for perm, val in buckets.items()}
```

```python
fig, ax = plt.subplots(figsize =(12, 7))
#plt.figure(figsize=(10, 5))
# Add x, y gridlines
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.75,
        alpha = 0.75)
ax.barh([key.replace("_", " ") for key in total_counts.keys()], total_counts.values(), align = "center")
fontsize = 20
plt.xlim([0, 9000000])
plt.title("Permissions Count by Category", fontdict={'fontsize': fontsize})
#plt.xticks(rotation = 70)
for i in ax.patches:
    plt.text(i.get_width()+100000, i.get_y()+0.5,
             '{:,}'.format(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')
ax.get_xaxis().set_major_formatter(
    m.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
plt.yticks(fontsize=13)
plt.xticks(fontsize=10)
plt.savefig(fname = "permissions_category_all" + ".png", format = "png", bbox_inches = 'tight')
plt.show()

```
